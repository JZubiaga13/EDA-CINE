# I. INTRODUCCIÓN

En el mundo del cine, las películas no solo cuentan historias, sino que también representan un complejo entramado de decisiones financieras y artísticas. El análisis de datos se ha convertido en una herramienta poderosa para comprender estas dinámicas, proporcionando perspectivas útiles sobre los factores que determinan el éxito de una película, los cambios en las tendencias cinematográficas y las decisiones de reconocimiento por parte de las academias de cine. Este trabajo exploratorio se centra en analizar múltiples aspectos de la industria cinematográfica, utilizando datos relacionados con presupuestos, recaudación, premios y características de las películas.

El estudio aborda cuatro aspectos principales:

- La relación entre presupuesto y recaudación.
- Relación entre la duración de las películas, su recaudación y la evolución de la duración media a lo largo del tiempo.
- La tendencia hacia precuelas, secuelas, remakes y reboots en detrimento de guiones originales.

## CONTEXTO
La industria cinematográfica ha experimentado una transformación profunda desde sus inicios en la década de 1920. Durante esta época, conocida como la Edad de Oro de Hollywood, los grandes estudios dominaron la producción y distribución de películas, estableciendo géneros clásicos y un sistema que marcó la pauta para la evolución en décadas posteriores. A medida que el tiempo avanzaba, la industria comenzó a diversificarse, adoptando tecnologías como el sonido, el color y los efectos digitales, cada uno de los cuales redefinió las posibilidades narrativas y visuales del cine.

En las últimas décadas, la globalización han acelerado aún mas estos cambios. Las plataformas de streaming han alterado profundamente los hábitos de consumo, permitiendo que las audiencias accedan a un catálogo casi ilimitado de películas desde cualquier lugar del mundo. Sin embargo, estos avances también han planteado nuevos desafíos, especialmente en términos de competencia por la atención del público y el modelo de negocio tradicional de las salas de cine.

El impacto de la pandemia de COVID-19 en 2020 marcó un punto de inflexión en la industria. Los cierres de salas de cine, los retrasos en las producciones y la dependencia creciente del streaming cambiaron el panorama para siempre. Muchas películas fueron lanzadas directamente en plataformas digitales, mientras que los cines enfrentaron una crisis sin precedentes, reduciendo su afluencia de público incluso después del levantamiento de las restricciones. Este periodo también incentivó a los estudios a apostar más por franquicias conocidas y formatos seguros, como remakes, secuelas y reboots, en un intento por mitigar riesgos financieros.

A esto se sumaron las huelgas de guionistas y actores en 2022 y 2023, que reflejaron un descontento generalizado por las condiciones laborales, la distribución de ingresos en la era del streaming y el uso de tecnologías como la inteligencia artificial. Estas huelgas paralizaron temporalmente grandes sectores de la industria, destacando la tensión entre la creatividad artística y las fuerzas económicas que impulsan la producción cinematográfica.

Con todo esto se expone la necesidad de comprender cómo las dinámicas económicas, sociales y tecnológicas influyen en el cine, tanto en términos de su creación como de su recepción por parte de nosotros los consumidores del septimo arte.



## OBJETIVOS E HIPÓTESIS

**Relación entre presupuesto y recaudación**
Se parte de la idea de que el presupuesto de una película influye directamente en su capacidad de generar ingresos. Esto se debe a que mayores presupuestos suelen implicar mejores efectos visuales, reparto de renombre y campañas de marketing más amplias. La hipótesis plantea que a <u>**mayor presupuesto, mayor será la recaudación en taquilla**.</u>

**Duración de las películas y su recaudación**
Considero que, en líneas generales, las películas tienen una duración óptima en términos de aceptación del público, y esta suele situarse en torno a las dos horas. Cuando se piensa en una película, considero que la producción que se crea en la cabeza de cada uno ronda esa duración. La primera hipótesis en este ámbito sugiere que <u>**las películas con duraciones cercanas a las dos horas tienden a recaudar más**</u>, mientras que aquellas que se alejan significativamente, tanto hacia un extremo como hacia el otro, recaudan menos de media.Por otro lado, se explorará la evolución de la duración media de las películas a lo largo de los años para identificar la tendencia y tratar de responder a la *preocupación social* de que pronto <u>**las películas superarán las tres horas de duración con más frecuencia**</u>.

**Incremento de guiones repetidos frente a ideas originales**
En los últimos años, se ha observado un aumento en la producción de precuelas, secuelas, remakes y reboots, muchas veces priorizados por su capacidad de atraer a una audiencia ya cautiva. La hipótesis plantea que desde la pandemia aproximadamente, <u>**la industria cinematográfica ha reducido la producción de películas con ideas originales en favor de estos formatos repetidos o adaptados**</u>. Cabe mencionar que no es una busqueda de criticar a la industria ni a los productores y guionistas de Hollywood, sino un acercamiento a como los eventos recientes como el COVID o las huelgas han podido incidir en la búsqueda de proyectos más sencillos o menos arriesgados.

**Género y premios a mejor película**
Los premios a mejor película suelen estar influenciados por tendencias de género y estilo. La hipótesis considera que <u>**las películas del género drama reciben un número significativamente mayor de premios Oscar y Goya a mejor película en comparación con otros géneros**</u> como la comedia o la acción.

Con este análisis, se espera no solo validar o refutar estas hipótesis, sino también aportar una visión más profunda sobre las dinámicas de la industria cinematográfica y cómo éstas impactan en el arte, los negocios y el reconocimiento público.

Para las conclusiones voy a estructurarme en dos apartados principales:

### ANÁLISIS DE LAS VISUALIZACIONES OBTENIDAS CON RESPECTO A LAS HIPÓTESIS

**Relación entre presupuesto y recaudación**

Los gráficos que analizan esta relación muestran una tendencia general positiva entre el presupuesto y la recaudación, con una línea de regresión que roza los 45 grados. Las producciones con presupuestos más altos suelen generar ingresos significativamente mayores en taquilla, aunque hay una dispersión dentro de cada rango presupuestario y se aprecian muchas excepciones, lo que puede atribuirse al efecto de las grandes campañas de marketing y al impacto de nombres reconocidos tanto en el elenco como en la dirección. Un aspecto a destacar es que los grandes presupuestos se concentran en las producciones del siglo XXI, donde los estudios y productoras han alcanzado su máxima expansión, y aprovechan cada mínimo éxito en taquilla para futuras producciones y campañas. Por otro lado, también se observan un número considerable de grandes producciones que no logran un retorno significativo y en el otro sentido, películas de mediados del siglo pasado que con presupuestos mas humildes, lograron recaudar cifras extraordinarias. Lo que indica que un gran presupuesto no garantiza con absoluta eficacia un éxito absoluto, pero si un resultado que supera, salvo escasas excepciones, la recaudación media.

**Duración de las películas y su recaudación**

Los datos brutos indican que, como norma general, las películas con duraciones cercanas a las dos horas tienden a obtener buenas recaudaciones, apoyando parcialmente la hipótesis de que esta es una duración óptima. Sin embargo, los gráficos muestran que la rentabilidad media más alta corresponde a películas de más de tres horas. Aunque estas películas representan una minoría en comparación con las de dos horas, su desempeño sugiere que, bajo ciertas circunstancias, las narrativas largas pueden ser muy exitosas. 
Otro hallazgo relevante es que las películas más largas presentan menos casos de baja recaudación, mientras que aquellas entre una y dos horas abarcan un rango amplio, desde grandes éxitos hasta fracasos. Por último, aunque hay un leve aumento en la duración media desde 1950, en los últimos 50 años esta se ha mantenido relativamente estable, alrededor de los 110 minutos.

Respecto de la hipótesis sobre las cuatro horas, se ha analizado la duración de la película más larga de cada año, obteniendo una conclusión negativa, pues no se aprecia una tendencia creciente clara, rondando desde hace décadas los 190 minutos. El futuro de esta hipótesis depende, claro está, depende de la recepción de las películas que superan esta cifra por parte de nosotros los consumidores. 

**Incremento de guiones repetidos frente a ideas originales**
Contrariamente a lo esperado, los gráficos no muestran un aumento significativo en la producción de secuelas, precuelas, remakes y reboots desde el año 2020. Según lo obtenido, la mayoría de estas producciones se concentran entre los 2000 y 2010. No obstante, este resultado puede estar influido por las limitaciones del análisis, que solo cuenta con películas hasta 2024, y se basó en aspectos medibles como la repetición de palabras en títulos y descripciones. Pese a todo, a nivel de percepción pública parece claro que desde 2015-2020 ha aumentado la reutilización de guiones e ideas existentes, con una menor proporción de ideas originales. 

**Género y premios a mejor película**

El análisis de los premios Óscar y Goya confirma que el drama es con gran diferencia el género predominante en las categorías de mejor película. Aunque otros géneros como la comedia o la acción también están presentes, su frecuencia es considerablemente menor. Esto sugiere una inclinación crítica hacia historias que abordan temas humanos, serios y emotivos, en lugar de narrativas ligeras o centradas en el entretenimiento.

### LOS RESULTADOS Y SUS IMPLICACIONES PARA LA INDUSTRIA CINEMATOGRÁFICA

**Presupuesto y recaudación: el equilibrio casi perfecto entre inversión y retorno**

La relación positiva entre presupuesto y recaudación subraya la importancia de los grandes estudios en el panorama cinematográfico actual. Sin embargo, también resalta la volatilidad del mercado: un presupuesto elevado no garantiza el éxito, y un presupuesto modesto no significa un fracaso. Este hallazgo enfatiza la necesidad de estrategias de mercado y narrativa bien definidas. Para estudios independientes y mercados más pequeños, estas conclusiones sugieren que competir con Hollywood en términos de presupuesto es insostenible, lo que los lleva a buscar nichos o enfoques creativos que destaquen con menos recursos.

**Duración de las películas y su impacto en la experiencia del espectador**

El hallazgo de que las películas cercanas a las dos horas tienen mejor desempeño apoya la idea de que existe una duración óptima para maximizar la aceptación del público. La tendencia al alza en la duración media plantea un reto: mientras algunas audiencias aprecian narrativas más desarrolladas, otras pueden sentirse alienadas por películas excesivamente largas. La industria debe equilibrar esta evolución con la atención al público, particularmente en un contexto donde la competencia por la atención del público no deja de crecer, con la expansión de las plataformas de streaming, los videojuegos y las redes sociales. Estas últimas, ademas, han incentivado una reducción cada vez mas visible en la capacidad de concentración y atención duradera de las nuevas generaciones, lo que supone un obstáculo contra esas películas excesivamente largas. Por todo ello se interpreta que la producción de una película que roce o supere las tres horas requiere una profunda comprensión previa de las preferencias del público si se quiere obtener una buena taquilla.

**Ideas originales vs. guiones repetidos: ¿creatividad en peligro?**

Aunque los limitados gráficos no acompañan a la hipótesis, si muestran que en el siglo XXI la reutilización y adaptación de guiones y temas ya existentes se ha disparado, y el la opinión pública generalizada así lo reflejan. Esto indica que la evolución de la industria, sumado a eventos como las huelgas y la pandemia, incentivaron proyectos menos arriesgados y más dependientes de franquicias conocidas. Esto ha permitido la continuidad de producciones durante momentos de incertidumbre, pero plantea preocupaciones sobre la diversidad narrativa. La apuesta por franquicias y formatos conocidos puede limitar la innovación, alienando a audiencias que buscan propuestas frescas. 

Este hallazgo sugiere que la tensión entre creatividad y seguridad financiera continua: Si bien los formatos repetidos pueden garantizar ingresos, una dependencia excesiva de estos podría erosionar la creatividad y la lealtad de ciertas audiencias y cansar al espectador medio. Para mantener la relevancia a largo plazo, los estudios deben equilibrar el riesgo con la innovación narrativa.

**Géneros y premios: la validación de los dramas**

La predominancia del drama en premios como el Oscar y el Goya sugiere que los críticos y jurados valoran las historias que exploran temas profundos y humanos. Esto influye directamente en la percepción pública del cine de calidad, reforzando la idea de que el drama es el género más "prestigioso". Para otros géneros, esto implica una mayor dificultad para obtener reconocimiento crítico, a pesar de su popularidad entre el público.

**Transformación del modelo de negocio**

El auge de los servicios de streaming ha reducido la dependencia de la taquilla tradicional. Los estudios deben adaptar sus estrategias, considerando no solo la producción sino también la distribución en plataformas digitales. La pandemia aceleró esta transición, inclinando la balanza claramente hacia la comodidad del hogar y subrayando la necesidad de un replanteamiento por parte de los estudios de cada estrategia, cada producción y cada idea que se quiera llevar a la pantalla grande. 

Así, y aunque desde la vuelta a la normalidad las salas de cine han mostrado una lenta pero efectiva resiliencia, su modelo de negocio enfrenta presiones significativas. La industria debe replantear la experiencia teatral desde sus fundamentos, ofreciendo un valor agregado que no pueda replicarse en casa.



**En síntesis, el EDA aquí desarrollado confirma que la industria cinematográfica está en un momento de cambio acelerado, con importantes retos y oportunidades. La adaptación será clave para garantizar su relevancia y sostenibilidad en el futuro.**
